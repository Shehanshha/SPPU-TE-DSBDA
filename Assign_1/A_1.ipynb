# Data preprocessing:
# check Missing values

```python
df.isnull()
```

```python
df.isnull().sum() #when data is qualitative the calculate mode of this and replace by mode
```

```python
df1 = df["Age"].mean()
df1
df["Age"].replace(np.nan,df1,inplace = True)
```

```python
df2 = df["Cabin"].mode()
df2
```

```python
df["Cabin"].replace("C23",inplace = True)
```

```python
df["Embarked"].fillna("S",inplace = True)
```

```python
df.isnull().sum()
```

# Data Formatting and Data Normalization:

```python
#use dropna() for removing NAN values from dataset
df = df.dropna()
```

```python
var = df.dtypes
print(var)
```

```python
col = df.select_dtypes(include=['object']).columns
col
```

```python
col = df.select_dtypes(include=['int64']).columns
col
```

```python
#type conversions.
df['Age'] = df['Age'].astype('int64')
df['Age'].dtype
```

#  Turn categorical variables into quantitative

```python
df["Sex"].replace({'female':0,"male":1},inplace = True)
df
```

```python
#method2
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Apply LabelEncoder to the categorical column
df["Sex"] = label_encoder.fit_transform(df["Sex"])
df['Sex']
```

```python
# method2
df2 = pd.get_dummies(df['Sex'],prefix = 'Category')
df2
```
