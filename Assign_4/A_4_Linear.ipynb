```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error  # for chaking the result
```

```python
df = pd.read_csv('BostonHousing.csv')
df
```

```python
df['rad'].value_counts()
```

```python
df.columns
```

```python
# divide the into x and y 
# x is independent variable and y is depended
x = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b', 'lstat']]
y=df['medv']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=37)  #test_size=0.2 specifies that 20% of the data will be used for testing.
```

```python
#traning a model
model = LinearRegression() #create a object
model.fit(x_train,y_train) #fit is used for finding the relation between them and fit
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
data = pd.DataFrame({"actual":y_test,"predicted":y_predict})
data
```

```python
# traning accuracy
model.score(x_train,y_train)*100
```

```python
#testing accuracy
model.score(x_test,y_test)*100
```

```python
# check error 
mean_squr_error = np.sqrt(mean_squared_error(y_test,y_predict))
mean_squr_error
```
