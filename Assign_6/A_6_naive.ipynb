```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gau = GaussianNB()
gau.fit(x_train,y_train)
```

```python
y_predict = gau.predict(x_test)
y_predict
```

```python
data = pd.DataFrame({"actual":y_test,"predicted":y_predict})
data
```

```python
gau.score(x_train,y_train)*100
```

```python
gau.score(x,y)*100
```

```python
cm = confusion_matrix(y_test,y_predict)
cm
```

# Accuracy


```python
a = accuracy_score(y_test,y_predict)
a
```

# Error rate

```python
e = 1 - a
e
```

# Precision

```python
p = precision_score(y_test,y_predict,average='macro')  # average='micro' used to calculate the precision metric for a multiclass classification problem by averaging the precision of each class.
p
```

# Recall


```python
r = recall_score(y_test,y_predict,average='macro')
r
```

# F-Sore

```python
f = 2*(p*r)/(p+r)
f
```

```python
TP = cm[0][0] + cm[1][1] + cm[2][2]
FP = cm[0][1] + cm[0][2] + cm[1][0] + cm[1][2] + cm[2][0] + cm[2][1]
```

```python
a1 = (TP)/(TP+FP)
a1
```
